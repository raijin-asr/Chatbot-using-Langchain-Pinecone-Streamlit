```python
embeddings = OpenAIEmbeddings(
  model= "text-embedding-3-small"
)
index_name= "chatbot2"

#split documents into chunks
text_splitter = RecursiveCharacterTextSplitter()
split_docs = text_splitter.split_documents(docs)
```
