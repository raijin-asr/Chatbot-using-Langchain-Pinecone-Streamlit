```python
from langchain_community.document_loaders import DirectoryLoader
import glob
import fitz

directory = '/content/data'

def load_docs(directory):
  loader = DirectoryLoader(directory)
  documents = []
  for filename in glob.glob(f"{directory}/**/*.pdf"):  # Use glob to find PDFs
    print(f"Found PDF file: {filename}")  # Debug: Print found filenames
    try:
      with fitz.open(filename) as doc:
        text = ""
        for page in doc:
          text += page.get_text()  # Extract text from each page
        documents.append({"content": text})
      print(f"Successfully extracted text from {filename}")  # Debug: Print successful extraction
    except Exception as e:
      print(f"Error processing {filename}: {e}")  # Debug: Print any errors

  return documents

documents = load_docs(directory)

print(f"Number of documents loaded: {len(documents)}")
```

# 2. Splitting documents

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_docs(documents,chunk_size=500,chunk_overlap=20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs

docs = split_docs(documents)
print(len(docs))
```

# 3. Creating embeddings

```python
from langchain.embeddings import SentenceTransformerEmbeddings
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
query_result = embeddings.embed_query("Welcome to Chatbot")
len(query_result) # print the length of the embedding vector
```

# 5. Access and search embeddings using the similarity_search

```python
def get_similiar_docs(query,k=1,score=False):
  if score:
    similar_docs = index.similarity_search_with_score(query,k=k)
  else:
    similar_docs = index.similarity_search(query,k=k)
  return similar_docs

query = "How is Nepal economy"
similar_docs = get_similiar_docs(query)
similar_docs
```
